```python
import pandas as pd
import numpy as np
```

```python
matches = pd.read_csv("epl_matches_train.csv")
```

```python
matches = matches[matches["season"] == "2014/2015"]
```

```python
matches
```

```python
matches['match_id'].to_csv("matches2014_2015_id.csv", index = False)
```

```python
home_unique_values = matches["home_team_id"].unique()
away_unique_values = matches["away_team_id"].unique()

unique_values = set(np.concatenate((home_unique_values, away_unique_values)))
unique_values
```

```python
results = {'id': list(unique_values)}
df = pd.DataFrame(results)
df
```

```python
df.shape
```

```python
df["score"] = 0.0

home_wins = matches[matches["home_team_goal"] > matches["away_team_goal"]].groupby("home_team_id").size()
home_draws = matches[matches["home_team_goal"] == matches["away_team_goal"]].groupby("home_team_id").size() * 0.5

away_wins = matches[matches["away_team_goal"] > matches["home_team_goal"]].groupby("away_team_id").size()
away_draws = matches[matches["away_team_goal"] == matches["home_team_goal"]].groupby("away_team_id").size() * 0.5
team_scores = home_wins.add(home_draws, fill_value=0).add(away_wins, fill_value=0).add(away_draws, fill_value=0)

df["score"] = df["id"].map(team_scores).fillna(0)
```

```python
df
```

```python
df.shape
```

```python
df = df.sort_values("score", ascending=False)
```

```python
df
```

```python
df["win_probability"] = round(df["score"] / max(df["score"]), 2)
```
