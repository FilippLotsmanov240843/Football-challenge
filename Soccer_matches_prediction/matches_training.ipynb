```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
players = pd.read_csv('epl_players.csv')
matches_train = pd.read_csv('epl_matches_train.csv')
```

```python
matches_train
```

```python
matches_train = matches_train.drop_duplicates()
```

```python
player_cols = [col for col in matches_train.columns if "player" in col]
matches_train = matches_train.drop(columns=player_cols, errors="ignore")
```

```python
matches_train = matches_train.dropna(subset=["home_team_goal", "away_team_goal"])
```

```python
matches_train.describe()
```

```python
matches_train["date"] = pd.to_datetime(matches_train["date"], errors='coerce')
```

```python
matches_train.isna().any()
```

```python
matches_train = matches_train.dropna()
```

```python
matches_train
```

```python
matches_train.to_csv('cleaned_training_matches.csv', index=False)
```
