```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
players = pd.read_csv('epl_players.csv')
```

```python
for col in players.columns:
    print(f" {col} has missing values: {players[col].isna().any()}")
```

```python
players = players[["player_id", "finishing", "dribbling", "acceleration", "sprint_speed", "positioning", "shot_power", "short_passing", "long_passing", "vision", "dribbling", "crossing", "ball_control", "marking", "standing_tackle", "sliding_tackle", "interceptions", "gk_diving", "gk_handling", "gk_kicking", "gk_positioning", "gk_reflexes", "date"]]
```

```python
players.shape
```

```python
players = players.dropna()
```

```python
players.shape
```

```python
players['date'] = pd.to_datetime(players['date'], errors='coerce')

# Filter dataset to include only players with records in 2015 or December 2014
players = players[(players['date'].dt.year == 2016) |
                           ((players['date'].dt.year == 2015) & (players['date'].dt.month == 12))]

# Sort by latest date first to keep the most recent record per player
players = players.sort_values(by=['player_id', 'date'], ascending=[True, False])

# Keep only the latest record per player
players = players.drop_duplicates(subset=['player_id'], keep='first')

# Reset index for clarity
players.reset_index(drop=True, inplace=True)
```

```python
players
```

```python
attacker_cols = ["finishing", "dribbling", "acceleration", "sprint_speed", "positioning", "shot_power"]
midfielder_cols = ["short_passing", "long_passing", "vision", "dribbling", "crossing", "ball_control"]
defender_cols = ["marking", "standing_tackle", "sliding_tackle", "interceptions"]
goalkeeper_cols = ["gk_diving", "gk_handling", "gk_kicking", "gk_positioning", "gk_reflexes"]

# Adjusted function to create a single FIFA rating column based on the highest attribute category
def calculate_fifa_rating(df):
    # Compute average scores for each player category
    df["Attacker_Score"] = df[attacker_cols].mean(axis=1, skipna=True)
    df["Midfielder_Score"] = df[midfielder_cols].mean(axis=1, skipna=True)
    df["Defender_Score"] = df[defender_cols].mean(axis=1, skipna=True)
    df["Goalkeeper_Score"] = df[goalkeeper_cols].mean(axis=1, skipna=True)

    # Identify the highest role for each player
    df["Best_Role"] = df[["Attacker_Score", "Midfielder_Score", "Defender_Score", "Goalkeeper_Score"]].idxmax(axis=1)

    # Assign FIFA rating based on the highest role's score
    df["FIFA_rating"] = df.apply(lambda x: x[x["Best_Role"]], axis=1)
    df['Position'] = df["Best_Role"].str.replace('_Score$', '', regex=True)

    # Remove helper columns
    df.drop(columns=["Attacker_Score", "Midfielder_Score", "Defender_Score", "Goalkeeper_Score", "Best_Role"], inplace=True)

    return df

# Apply the function to the players DataFrame
players = calculate_fifa_rating(players)

players
```

```python
players.shape
```

```python
players = players.drop_duplicates()
```

```python
players.shape
```

```python
players = players.sort_values(by=["FIFA_rating"], ascending=[False])
players['FIFA_rating'] = players['FIFA_rating'].astype(int)
```
