```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
```

```python
teams = pd.read_csv('epl_teams.csv')
```

```python
teams.head()
```

```python
teams["date"] = pd.to_datetime(teams["date"], errors='coerce')
```

```python
teams = teams.sort_values(by="date").drop(columns=["date"], errors='ignore').drop_duplicates(subset=["team_id"], keep="last")
```

```python
teams = teams.dropna()
```

```python
teams.to_csv('cleaned_teams.csv', index=False)
```
